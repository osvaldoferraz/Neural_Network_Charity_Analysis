### Deliverable 1: Preprocessing the Data for a Neural Network

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

### OPTIMIZATION - Droping two more columns

```python
#Check values on STATUS
application_df.STATUS.value_counts()
```

```python
#Check values on SPECIAL_CONSIDERATONS
application_df.SPECIAL_CONSIDERATIONS.value_counts()
```

```python
# Drop the non-beneficial ID columns, 'EIN','NAME','STATUS','SPECIAL_CONSIDERATIONS'.
application_df = application_df.drop(columns=['EIN', 'NAME','STATUS','SPECIAL_CONSIDERATIONS'])
application_df
```

```python
# Determine the number of unique values in each column.
unique_values = application_df.nunique()
unique_values
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_type = application_df.APPLICATION_TYPE.value_counts()
app_type
```

```python
# Visualize the value counts of APPLICATION_TYPE
app_type.plot.density()
```

```python
# Determine which values to replace if counts are less than ...?
replace_app = list(app_type[app_type < 500].index)
replace_app

# Replace in dataframe
for app in replace_app:
    application_df.APPLICATION_TYPE = application_df.APPLICATION_TYPE.replace(app,"Other")
    
# Check to make sure binning was successful
application_df.APPLICATION_TYPE.value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification = application_df.CLASSIFICATION.value_counts()
classification
```

```python
# Visualize the value counts of CLASSIFICATION
classification.plot.density()
```

```python
# Determine which values to replace if counts are less than ..?
replace_class = list(classification[classification < 1800].index)
replace_class

# Replace in dataframe
for cls in replace_class:
    application_df.CLASSIFICATION = application_df.CLASSIFICATION.replace(cls,"Other")
    
# Check to make sure binning was successful
application_df.CLASSIFICATION.value_counts()
```

```python
# Generate our categorical variable lists
app_cat = application_df.dtypes[application_df.dtypes == 'object'].index.tolist()
app_cat
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(application_df[app_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(app_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
application_df = application_df.merge(encode_df, left_index=True, right_index=True)
application_df = application_df.drop(app_cat, 1)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df.IS_SUCCESSFUL
X = application_df.drop(columns=['IS_SUCCESSFUL'])

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Deliverable 2: Compile, Train and Evaluate the Model

### Optimization - Increase hidden layers

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 30
hidden_nodes_layer3 = 20

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, 
                          input_dim=number_input_features, 
                          activation="relu")
)

# Second hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer2, 
                          activation="relu")
)

# Third hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer3, 
                          activation="relu")
)

# Output layer
nn.add(
    tf.keras.layers.Dense(units=1, 
                          activation="sigmoid")
)

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# import checkpoint dependencies
import os
from tensorflow.keras.callbacks import ModelCheckpoint

#Define the checkpoint path and filenames
os.makedirs('checkpoints/optmization/', exist_ok=True)
checkpoint_path = 'checkpoints/optmization/weights.{epoch:02d}.hdf5'
# Create a callback that sves the model's weights every 5 epochs
cp_callback = ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    save_weights_only=True,
    save_freq=3500
)
```

```python
# Train the model
#fit_model = nn.fit(X_train_scaled, y_train, epochs=50, callbacks=[cp_callback]) 
```

### OPTIMIZATION - Increase # of Epochs

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, callbacks=[cp_callback]) 
```

```python
# Plot the loss function
import matplotlib.pyplot as plt

plt.plot(fit_model.history["loss"])
```

```python
# Plot the accuracy function
import matplotlib.pyplot as plt

plt.plot(fit_model.history["accuracy"])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimization.h5")
```
